```python
import torch
from tinyyolov2 import TinyYoloV2
from pruned_tinyyolov2NoBN import PrunedTinyYoloV2NoBN
from tinyyolov2NoBN import TinyYoloV2NoBN
```

```python
input = torch.randn(1,3,320,320).to(dtype=torch.float8)

model = PrunedTinyYoloV2NoBN(num_classes = 1)
sd = torch.load("models/voc_pruned_r002_1_times_19_epochs_lr0001_decay0005.pt")
model.load_state_dict(sd)
model.eval()
model.to(dtype=torch.float8)
torch.onnx.export(model, input, "onnx/quant_test.onnx", verbose=True, opset_version=10)
```
